## **Execution Time and Cost for Instances and Number of VMs per each Instance**

```python
vmInstanceName = np.array(["t4g.small", "t4g.micro", "t3.medium"])
numberofinstances = np.array([148, 107, 167])
totalExecTime = np.array([2838895.0, 835382.0, 5134573.816])
totalCost = np.array([16.7179372, 2.4597359, 75.3070826])
execTimeInHours = totalExecTime / 3600

# Calculate total execution time in hours and total cost
totalExecTimeInHours = np.sum(execTimeInHours)
totalCostSum = np.sum(totalCost)

# Create a figure with 2 side-by-side plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 6))

# Plot for execution time
bars1 = ax1.bar(vmInstanceName, totalExecTime, color='skyblue', edgecolor='black')
ax1.set_xlabel("Instance Names")
ax1.set_ylabel("Execution Time in Seconds")
ax1.set_title("Execution Time per Instance")
ax1.set_xticks(np.arange(len(vmInstanceName)))
ax1.set_xticklabels(vmInstanceName)

# Add the execution time and number of instances on top of the bars
for i, bar in enumerate(bars1):
    ax1.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f'{execTimeInHours[i]:.2f}hrs\n({numberofinstances[i]} instances)', ha='center', va='bottom', fontsize=10)

# Add the total execution time (in hours) on the top-left corner
ax1.text(-0.3, 0.9 * max(totalExecTime), f'Total Execution Time: {totalExecTimeInHours:.2f} hrs',
         ha='left', va='bottom', fontsize=12, fontweight='bold',
         bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))

# Plot for cost
bars2 = ax2.bar(vmInstanceName, totalCost, color='lightgreen', edgecolor='black')
ax2.set_xlabel("Instance Names")
ax2.set_ylabel("Total Cost")
ax2.set_title("Cost per Instance")
ax2.set_xticks(np.arange(len(vmInstanceName)))
ax2.set_xticklabels(vmInstanceName)

# Add the cost and number of instances on top of the bars
for i, bar in enumerate(bars2):
    ax2.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f'${totalCost[i]:.2f}\n({numberofinstances[i]} VM Instances Under this VM)', ha='center', va='bottom', fontsize=10)

# Add the total cost on the top-left corner
ax2.text(-0.3, 0.9 * max(totalCost), f'Total Cost: ${totalCostSum:.2f}',
         ha='left', va='bottom', fontsize=12, fontweight='bold',
         bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
